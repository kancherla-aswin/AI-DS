3.count of each age group with individual count of age group wrt **gender**

4.plot the total amt spent by each age **group**

```python
import plotly.express as px

amt_by_age = diwali_df.groupby('Age Group')['Amount'].sum().reset_index()
# Use 'names' for labels and 'values' for slice sizes
px.pie(amt_by_age, names='Age Group', values='Amount', title='Total Amount Spent by Each Age Group', color="Amount")
```

5.plot the graph that depicts total number of orders from top 10 states

```python
import plotly.express as px

state_orders_df = diwali_df['State'].value_counts().nlargest(10).reset_index()
state_orders_df.rename(columns={'index': 'State', 'count': 'Orders'}, inplace=True)
# Use 'names' for labels and 'values' for slice sizes
fig = px.pie(state_orders_df, names='State', values='Orders', title='Total Number of Orders from Top 10 States',
             color='Orders')
fig.show()
```

6. Total amt of top 10 states

```python
state_amount = diwali_df.groupby('State')['Amount'].sum().reset_index(name='Total Amount')
print(state_amount)
```

```python
import plotly.express as px
tot_amt = px.bar(state_amount, x='State', y='Total Amount', title='Total Amount Spent by Each State',color="Total Amount")
tot_amt.show()
```

7.plot me a graph that shows comparison between no of married and unmarried people

```python
marital_counts = diwali_df['Marital_Status'].value_counts()
fig = px.bar(marital_counts, x=marital_counts.index, y=marital_counts.values)

# Set the width of the graph
fig.update_layout(width=500, height=700)

fig.show()
```

8.plot me amt spent by each male and each femlae wrt to marital status


```python
amt_by_marital_gender = diwali_df.groupby(['Gender', 'Marital_Status'])['Amount'].sum().reset_index()
px.bar(amt_by_marital_gender, x='Gender', y='Amount', color='Marital_Status',
       barmode='group', title='Amount Spent by Gender and Marital Status',
       facet_col='Marital_Status',height=500,width=700)
```

9.plot me the count of each occupation that is present in dataset


```python
import plotly.express as px

occupation_counts = diwali_df['Occupation'].value_counts()
fig = px.bar(occupation_counts, x=occupation_counts.index, y=occupation_counts.values,
             labels={'x':'Occupation', 'y':'Count'}, title='Count of Each Occupation')
fig.show()
```

10. Plot the amt spent by each occupation in descending order


```python
# Calculate amount spent by each occupation
amt_by_occupation = diwali_df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False).reset_index()

# Create a bar chart
fig = px.bar(amt_by_occupation, x='Occupation', y='Amount',color='Amount',
             title='Amount Spent by Each Occupation (Descending)')
fig.show()
```

11.plot a graph that provides statistics on the intrest of customers wrt to product category

```python
category_interest = diwali_df.groupby(['Product_Category'])['Orders'].count().reset_index()
px.bar(category_interest, x='Product_Category', y='Orders', title='Customer Interest by Product Category',color="Orders")
```

12.i would like to know the budget spent on each product category in descending oreder



```python
# Calculate budget spent on each product category
budget_by_category = diwali_df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False).reset_index()

# Print the result
print(budget_by_category)
```

```python
budget = px.bar(budget_by_category, x='Product_Category', y='Amount',
             title='Budget Spent on Each Product Category (Descending)',color="Amount")
budget.show()
```

```python
# Group data by age group and count orders
age_group_orders = diwali_df.groupby('Age Group')['Orders'].sum().reset_index()

# Create a bar chart using Plotly Express
import plotly.express as px
figf = px.pie(age_group_orders, names='Age Group', values='Orders', # Use 'names' for labels and 'values' for slice sizes
             title='Number of Orders by Age Group', color='Orders')
figf.show()
```

```python
# Group data by gender and product category, count orders
gender_category_orders = diwali_df.groupby(['Gender', 'Product_Category'])['Orders'].sum().reset_index()

# Create a grouped bar chart using Plotly Express
import plotly.express as px
fig1= px.bar(gender_category_orders, x='Product_Category', y='Orders', color='Gender',
             title='Product Category Purchases by Gender', barmode='group')
fig1.show()
```

```python
# Group data by marital status and occupation, count occurrences
marital_occupation_counts = diwali_df.groupby(['Marital_Status', 'Occupation'])['Orders'].count().reset_index()

# Create a grouped bar chart using Plotly Express
import plotly.express as px
fig3 = px.bar(marital_occupation_counts, x='Occupation', y='Orders', color='Marital_Status',
             title='Occupation Distribution by Marital Status',barmode='group',facet_col='Marital_Status')
fig3.show()
```

```python
# Group data by state and marital status, count occurrences
state_marital_counts = diwali_df.groupby(['State', 'Marital_Status'])['Orders'].count().reset_index()

# Create a grouped bar chart using Plotly Express
import plotly.express as px
fig4 = px.bar(state_marital_counts, x='State', y='Orders', color='Marital_Status',
             title='Marital Status Distribution by State', barmode='group',facet_col='Marital_Status')
fig4.show()
```

```python
# Group data by zone and occupation, count occurrences
zone_occupation_counts = diwali_df.groupby(['Zone', 'Occupation'])['Orders'].count().reset_index()

# Create a stacked bar chart using Plotly Express
import plotly.express as px
fig5 = px.bar(zone_occupation_counts, x='Zone', y='Orders', color='Occupation',
             title='Occupation Distribution by Zone', barmode='group')
fig5.show()
```

```python
# Group data by age group and marital status, count occurrences
age_marital_counts = diwali_df.groupby(['Age Group', 'Marital_Status'])['Orders'].count().reset_index()

# Create a stacked bar chart using Plotly Express
import plotly.express as px
fig5 = px.bar(age_marital_counts, x='Age Group', y='Orders', color='Marital_Status',
             title='Marital Status Distribution by Age Group', barmode='stack',facet_col='Marital_Status')
fig5.show()
```

```python
# Group data by age group and zone, count occurrences
age_zone_counts = diwali_df.groupby(['Age Group', 'Zone'])['Orders'].count().reset_index()

# Create a grouped bar chart using Plotly Express
import plotly.express as px
fig6 = px.bar(age_zone_counts, x='Zone', y='Orders', color='Age Group',
             title='Age Group Distribution by Zone', barmode='group')
fig6.show()
```
